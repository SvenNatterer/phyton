# Exercise 4

Create a DataFrame with missing values and demonstrate different methods
for handling them. Create the following DataFrame:

| City     | Jan_Temp | Feb_Temp | Mar_Temp |
|----------|----------|----------|----------|
| New York | 2        | NaN      | 8        |
| London   | 7        | 9        | NaN      |
| Berlin   | 3        | 5        | 10       |
| Paris    | 6        | 8        | 12       |

Then perform the following operations: 1. Check which values are missing
using `isnull()` 2. Count the number of missing values in each column 3.
Fill missing values with the mean of each column 4. Drop rows that
contain any missing values

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "City": ["New York", "London", "Berlin", "Parins"],
    "Jan_Temp": [2,7,3,6],
    "Feb_Temp": [np.nan,9,5,8],
    "Mar_Temp": [8,np.nan,10,12]
})

print(df.ffill())
print(df.bfill())

print(df.isnull())
print(df.isnull().sum())
print(df.fillna(df.mean(numeric_only=True)))
print(df.dropna())
```

# Exercise 6

Create two DataFrames and perform merge operations. Create the following
data:

**Rappers DataFrame:**

| rapper_id | name           | city         |
|-----------|----------------|--------------|
| 1         | Drake          | Toronto      |
| 2         | Kendrick Lamar | Compton      |
| 3         | J. Cole        | Fayetteville |
| 4         | Travis Scott   | Houston      |

**Albums DataFrame:**

| album_id | rapper_id | album_name              | release_year |
|----------|-----------|-------------------------|--------------|
| 101      | 1         | Take Care               | 2011         |
| 102      | 2         | good kid, m.A.A.d city  | 2012         |
| 103      | 1         | Views                   | 2016         |
| 104      | 3         | 2014 Forest Hills Drive | 2014         |
| 105      | 5         | Astroworld              | 2018         |

Then perform the following merge operations:

1.  Inner merge (only rappers with albums)
2.  Left merge (all rappers, with album info where available)
3.  Right merge (all albums, with rapper info where available)
4.  Outer merge (all rappers and all albums)

```python
import pandas as pd

# Rappers DataFrame
rappers_data = {
    'rapper_id': [1, 2, 3, 4],
    'name': ['Drake', 'Kendrick Lamar', 'J. Cole', 'Travis Scott'],
    'city': ['Toronto', 'Compton', 'Fayetteville', 'Houston']
}
rappers_df = pd.DataFrame(rappers_data)

# Albums DataFrame
albums_data = {
    'album_id': [101, 102, 103, 104, 105],
    'rapper_id': [1, 2, 1, 3, 5],
    'album_name': [
        'Take Care',
        'good kid, m.A.A.d city',
        'Views',
        '2014 Forest Hills Drive',
        'Astroworld'
    ],
    'release_year': [2011, 2012, 2016, 2014, 2018]
}
albums_df = pd.DataFrame(albums_data)

#pd.merge(rappers_df, albums_df, on="rapper_id")
#pd.merge(rappers_df, albums_df, on="rapper_id", how="left")
#pd.merge(rappers_df, albums_df, on="rapper_id", how="right")
pd.merge(rappers_df, albums_df, on="rapper_id", how="outer")
```
